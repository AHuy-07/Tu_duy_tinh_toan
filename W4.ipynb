```python
#W4A8
def reverse(n):
  x = 0
  while(n > 0):
    x *= 10
    x += n % 10
    n //= 10
  return x
def chk(n):
  s = str(n)
  for i in range(len(s)//2):
    if(s[i] != s[len(s) - i - 1]):
      return False
  return True
n = int(input())
cnt = 0
while(chk(n) == False):
  n = n + reverse(n)
  cnt += 1
print(f"{cnt} {n}")
```

```python
#W4A9
import math
def chkn(n):
  s = str(n)
  s1 = set(s)
  if(len(s) == len(s1)):
    return True
  return False
def chks(n):
  x = int(math.sqrt(n))
  return x*x == n
n = int(input())
for i in  range (1, n + 1):
  if(chkn(i) and chks(i)):
    print(i, end = " ")
```

```python
#W4A10
def steps (x):
  cnt = 1
  while x > 1:
    if x % 2 == 0: x /= 2
    else: x = 3 * x + 1
    cnt += 1
  return cnt
n = int(input())
res = -1
for i in range(1, n + 1):
  if(res < steps(i)):
    res = steps(i)
    ans = i
print(f"{ans} {res}")
```
