```python
#W5A1
a, b = map(int, input().split())
def max():
  if a > b: return a
  return b
print(max())
```

```python
#W5A2
a, b = map(int, input().split())
def swap():
  return b, a
print(swap())
```

```python
#W5A3
from math import sqrt
n = int(input())
def check():
  for i in range(2, int(sqrt(n)) + 1):
    if n % i == 0: return False
  return True
print(check())
```

```python
#W5A4
n = int(input())
def check():
  s = 0
  for i in range(1, n):
    if n % i == 0: s += i
    if s == n: return True
  return False
print(check())
```

```python
#W5A5
n = list(map(int, input().split()))
k = int(input())
for i in range(len(n)):
  if n[i] == k:
    print(i + 1)
    break
```

```python
#W5A6
n = int(input())
def check():
  res = 1
  for i in range(1, n+1):
    res *= i
  return res
print(check())
```

```python
#W5A7
s = input()
def cal(a, op, b):
  if op == '+': return a + b
  if op == '-': return a - b
  if op == '*': return a * b
  if op == '/': return a / b
for i in range(len(s)):
  if s[i].isdigit() == False and s[i] != ' ' and s[i] != '.':
    op = s[i]
    a = float(s[:i])
    b = float(s[i+1:])
print(f"{cal(a, op, b):.2f}")
```

```python
#W5A8
a, b = map(int, input().split())
a = a ^ b
cnt = 0
while a > 0:
  if a % 2 == 1:
    cnt+=1
  a//=2
print(cnt)
```
