```python
#W6A4
s = input()
res = {}
pair = s.split()
for i in pair:
  if ":" in i:
    k, v = i.split(":", 1)
    if k in res:
      res[k].append(v)
    else:
      res[k] = [v]
print(res)
```

```python
#W6A5
s = list(map(int, input().split()))
res = {"positives" : 0, "negatives" : 0, "zeros" : 0}
for i in s:
  if i < 0:
    res["negatives"]+= 1
  elif i > 0:
    res["positives"]+= 1
  elif i == 0:
    res["zeros"]+= 1
print(res)
```

```python
#W6A9
a = input()
b = input()
a = a.strip('{} ')
b = b.strip('{} ')
x = a.split(", ")
y = b.split(", ")
res = {}
for i in x + y:
  if i:
    k, v = i.split(":")
    k = k[1:-1]
    if k in res:
      res[k] += int(v)
    else:
      res[k] = int(v)
print(res)
```

```python
#W6A14
a1 = list(map(int, input().split()))
a2 = list(map(int, input().split()))
mp1 = {}
mp2 = {}
for i in a1:
    mp1[i] = mp1.get(i, 0) + 1
for i in a2:
    mp2[i] = mp2.get(i, 0) + 1
li = []
for key in mp1.keys():
    if key in mp2:
        li.append(key)
li.sort()
print(li)
```

```python
#W6A16
n, m = map(int, input().split())
li = []
for i in range(n):
    tmp = list(map(int, input().split()))
    li.append(tmp)
for i in range(n):
    for j in range(m):
        print(f"{li[i][j]:>4d}", end = " ")
    print(end = "\n")
```

```python
#W6A17
n = int(input())
m = n
li = []
for i in range(n):
    tmp = list(map(int, input().split()))
    li.append(tmp)
for i in range(n):
    for j in range(m):
        if i == j:
            print(li[i][j], end = " ")
print(end = "\n")
for i in range(n):
    for j in range(m):
        if i + j == n - 1:
            print(li[i][j], end = " ")
```

```python
#W6A18
n, m, k = map(int, input().split())
li = []
for i in range(n):
    tmp = list(map(int, input().split()))
    li.append(tmp)
cot = 0
for i in range(n):
    cot += li[i][k]
hang = 0
for i in range(m):
    hang += li[k][i]
print(f"Sum of col k: {cot}")
print(f"Sum of row k: {hang}")
```

```python
#W6A20
def solve(li, k):
    for i in range(len(li)):
        if li[i] == k:
            return i + 1
    return -1
arr = list(map(int, input().split()))
k = int(input())
print(solve(arr, k))
```
